### Making fancy plots with Matplotlib.pyplot
Let's read some test data.

```python
import pandas as pd
import matplotlib.pyplot as plt

#load data
data = {"T": [2.3, 2.9, 3.2, 2.4, 0.8, -0.3, -0.4, -0.6, -7.8, -12.4],
        "Tmax" : [4, 6, 4.1, 4.1, 3, 0.7, 0.2, 0, -0.5, -7],
        "Tmin" : [-0.7, 1, 0, 1, 0.4, -0.5, -1.8, -0.8, -10.1, -16.6],
        "RR" : [1.7, 1.4, 6.5, 0.9, 8, 4, 6.5, 2, 0, 0]}

#load index
index = ["1960-01-01", "1960-01-02", "1960-01-03", "1960-01-04", "1960-01-05", "1960-01-06", "1960-01-07", "1960-01-08", "1960-01-09", "1960-01-10"]

#transform index to date
index = pd.DatetimeIndex(index)

#create df
df = pd.DataFrame(data = data, index = index)
```

Let's take a look at the data:

```python
df
```

The data can be visualized in an easy way:

```python
plt.plot(df)
```

If the line type should be changed, a color indicator combined with a line style must be declared afterwards. When two or more plots are conducted right after one another, they will be printed in the same plot.

#### Possible combinations:

Line type:
- single line () or (-)
- dashed (--)
- dotted (o)

Color:
- red (r)
- green(g)
- blue (b)
- yellow (y)

```python
plt.plot(df["Tmax"], "r")
plt.plot(df["Tmin"], "b--")
```

When adding a legend, among others the location and the number of columns can be defined.

```python
#data
plt.plot(df["Tmin"], "b--", label = "Min temp")
plt.plot(df["Tmax"], "r", label = "Max temp")
#legend
plt.legend(loc = "best", ncol = 1)
```

The axis can be labeled as follows.

```python
#data
plt.plot(df["Tmin"], "b--", label = "Min temp")
plt.plot(df["Tmax"], "r", label = "Max temp")
#legend
plt.legend(loc = "best", ncol = 1)
#axis labels
plt.xlabel("Date")
plt.ylabel("C")
```

A more clean way, especially, when more complex plots should be printed, is using the figure and axis element.

The **subplot()** creates an empty plot.
The **figure** element is used as the actual plot.
While the **axis** element refers to the whole page. Here, multiple plots can be arranged (is goint to be relevant later), but for now, we just using the figure plot.  

```python
fig, ax = plt.subplots()
ax.plot(df["Tmin"], "b--", label = "Min temp")
ax.plot(df["Tmax"], "r", label = "Max temp")
ax.set(xlabel = "Date", ylabel = "C",
       title = "Temperature over time")
ax.legend()
```
